```python
# Logistic Regression
# Heart Attack Prediction
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/IEEE-CIS-GHRCE-62361B/MLBootcamp/master/Week2/heart.csv')

data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = data.corr()      # correlation
plt.figure(figsize=(20, 20))    # the plot area size
sns.heatmap(corr, annot=True)   # heatmap
```

```python
from sklearn.model_selection import train_test_split as tts

x, y = data.drop('target', axis=1), data.target
xTrain, xTest, yTrain, yTest = tts(x, y, test_size=0.2)

from sklearn.preprocessing import StandardScaler as SS
scaler = SS()
```

```python
xTrain, xTest = scaler.fit_transform(xTrain), scaler.fit_transform(xTest)
# preprocessing data using StandardScaler
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(xTrain, yTrain)
```

```python
yPred = lr.predict(xTest)
```

```python
accuracyScore = lr.score(xTest, yTest)
print(f'Accuracy = {accuracyScore * 100}')
```
